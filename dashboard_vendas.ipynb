```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Gera dataset de vendas fictício mas realista
np.random.seed(42)

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
         'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

dados = {
    'Mes': meses,
    'Vendas': [142, 158, 175, 163, 189, 201, 195, 210, 198, 225, 248, 312],
    'Meta':   [150, 160, 170, 170, 180, 190, 200, 210, 210, 220, 240, 300],
    'Clientes_Novos': [32, 38, 41, 35, 47, 52, 49, 55, 51, 58, 63, 78],
    'Regiao': ['Sul', 'Sul', 'Sudeste', 'Sudeste', 'Norte', 'Norte',
               'Sul', 'Sudeste', 'Norte', 'Sul', 'Sudeste', 'Norte']
}

df = pd.DataFrame(dados)
df.head() #só aparece os 5 primeiros
```

```python
# Dashboard com 4 gráficos interativos
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        'Vendas vs Meta por Mês',
        'Crescimento de Clientes Novos',
        'Vendas por Região',
        '% Atingimento da Meta'
    ),
    specs=[
        [{"type": "xy"}, {"type": "xy"}],
        [{"type": "pie"}, {"type": "xy"}]
    ]
)

# 1. Vendas vs Meta
fig.add_trace(go.Scatter(x=df['Mes'], y=df['Vendas'], name='Vendas',
              line=dict(color='#5c9ee0', width=3), mode='lines+markers'), row=1, col=1)
fig.add_trace(go.Scatter(x=df['Mes'], y=df['Meta'], name='Meta',
              line=dict(color='#e05c5c', width=2, dash='dash')), row=1, col=1)

# 2. Clientes novos
fig.add_trace(go.Bar(x=df['Mes'], y=df['Clientes_Novos'], name='Clientes Novos',
              marker_color='#5ce0a8'), row=1, col=2)

# 3. Vendas por região
vendas_regiao = df.groupby('Regiao')['Vendas'].sum().reset_index()
fig.add_trace(go.Pie(labels=vendas_regiao['Regiao'], values=vendas_regiao['Vendas'],
              name='Regiões'), row=2, col=1)

# 4. % atingimento da meta
df['Atingimento'] = (df['Vendas'] / df['Meta'] * 100).round(1)
fig.add_trace(go.Bar(x=df['Mes'], y=df['Atingimento'], name='% Meta',
              marker_color=df['Atingimento'].apply(
                  lambda x: '#5ce07a' if x >= 100 else '#e05c5c')), row=2, col=2)

fig.update_layout(
    title_text='Dashboard de Vendas 2024',
    title_font_size=20,
    height=700,
    showlegend=True
)

fig.write_html('dashboard_vendas.html')
fig.show()
```
